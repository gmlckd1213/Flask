# 로깅<br>

**프로그램이 작동할 때 발생하는 이벤트를 추적하는 행위**<br>
프로그램의 문제들을 파악하고 유지보수 하는데 사용되며, 로깅을 통해 발생한 에러를 추적할 수 있다.

## DEBUG < INFO < WARNING < ERROR < CRITIACL <br>
기본 로거 레벨 세팅은 WARNING 이기 때문에 설정 없이 INFO, DEBUG 를 출력할 수 없다.<br><br>
DEBUG : 상세한 정보 <br>
INFO : 일반적인 정보 <br>
WARNING : 예상치 못하거나 가까운 미래에 발생할 문제 <br>
ERROR : 에러 로그. 심각한 문제<br>
CRITIACL : 프로그램 자체가 실행되지 않을 수 있는 문제<br><br>


에러가 발생했을 때 Flask의 logger를 사용하여 에러를 확인 가능하다.<br>
Flask에서 기본적으로 logging을 제공한다.

# 로깅 구현<br>
app.py
```python
from flask import Flask,render_template

app = Flask(__name__)


# 404 에러를 제어하는 코드를 추가하세요.
    
@app.errorhandler(404)
def page_not_found(error):
    app.logger.error(error)
    return render_template("page_not_found.html")

@app.route('/')
def hello_elice():
    return "Hello Elice!"


if __name__ == '__main__':
    app.run()
```
